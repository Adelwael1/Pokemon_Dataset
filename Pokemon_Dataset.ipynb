```python
#Tidy data 
#create new data from pokemons data to explain melt nore easily
new_pok =pok
new_pok.head()
```

```python
# lets melt
# id_vars = what we do not wish to melt
# value_vars = what we want to melt
melted =pd.melt(frame = new_pok , id_vars='Name' , value_vars=['Attack','Defense'])
melted.head(10)
```

```python
#PIVOTING DATA
# Index is name
# I want to make that columns are variable
# Finally values in columns are value
melted.pivot(index='Name' , columns='variable' , values='value').head(10)
```

```python
#concatenate the data 
first_pok = pok.head()
last_pok = pok.tail()

conc_pok_row = pd.concat([first_pok,last_pok] , axis=0 , ignore_index = True)

conc_pok_row
```

```python
first_pok = pok['Attack'].head()
last_pok = pok['Defense'].head()

conc_pok_column =pd.concat([first_pok,last_pok], axis=1)
conc_pok_column
```

# 4- Data Visualization

```python
f, ax = plt.subplots(figsize=(18, 18))
sns.heatmap(pok2.corr(), annot=True, linewidths=0.5, fmt='.1f', ax=ax, cmap='coolwarm')
plt.show()
```

```python
# Line Plot
# color = color, label = label, linewidth = width of line, alpha = opacity, grid = grid, linestyle = sytle of line
pok.Speed.plot(kind = 'line', color = 'g',label = 'Speed',linewidth=1,alpha = 0.5,grid = True,linestyle = ':')
pok.Defense.plot(color = 'r',label = 'Defense',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')
plt.legend(loc='upper right')     # legend = puts label into plot
plt.xlabel('x axis')              # label = name of label
plt.ylabel('y axis')
plt.title('Line Plot')            # title = title of plot
plt.show()
```

```python
#relation between speed and defence 
#x= attack , y = defence 
pok.plot(kind='scatter' , x='Attack' , y='Defense',alpha=0.5 , color='red')
plt.xlabel('Attack')
plt.ylabel('Defence')
plt.title('relation between attack and defence')
```

```python
pok.Speed.plot(kind='hist', bins=50 , figsize=(12,12))
plt.show()
```
